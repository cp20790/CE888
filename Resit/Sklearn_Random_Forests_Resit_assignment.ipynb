```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("books.csv")
def sort_and_top_10(books_df):
    sorted_df = books_df.sort_values(by=["authors", "title"], ascending=False)
    return sorted_df.head(10)

authors_df = df.groupby(["authors", "title", "average_rating"],as_index=False)["ratings_count"].count()
authors_df.head()
graph_df = sort_and_top_10(authors_df)
g = sns.catplot(x="average_rating", y="title", kind="point", col="authors", data=graph_df, sharey=False)
```

# Section 2: Pages Count by Category

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("books.csv")
df = df[df['num_pages'] < 1000]
Juvenile = df[df['categories'] == 'Juvenile Fiction']
Comics = df[df['categories'] == 'Comics & Graphic Novels']
History = df[df['categories'] == 'History']
Philosophy = df[df['categories'] == 'Philosophy']

Juvenile = Juvenile.groupby(["categories", "authors", "num_pages"])["num_pages"].sum()
Comics = Comics.groupby(["categories", "authors", "num_pages"])["num_pages"].sum()
History = History.groupby(["categories", "authors", "num_pages"])["num_pages"].sum()
Philosophy = Philosophy.groupby(["categories", "authors", "num_pages"])["num_pages"].sum()
#ab =Juvenile+Comics+History+Philosophy

#x0 = sns.displot(ab, kde=True, bins=26)
x1 = sns.displot(Juvenile, kde=True, bins=26)
x2 = sns.displot(History, kde=True, bins=26)
x3 = sns.displot(Philosophy, kde=True, bins=26)
x4 = sns.displot(Comics, kde=True, bins=26)
```
